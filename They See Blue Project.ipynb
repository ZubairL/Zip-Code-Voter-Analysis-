```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('TSB.csv')  # reading in file. File contains TSB Data and The Congressional Zip Code data
```

```python
df.head()
```

```python
df.info()
```

```python
df.rename(columns = {"Unnamed: 6":"Cong Zip Codes"}, inplace = True)    #Changing column titles and taking first 5 characters of zip code
df['Cong Zip Codes'] =df['Cong Zip Codes'].str[:5]
```

```python
df['Cong Zip Codes'] = df['Cong Zip Codes'].replace(np.nan, 00000) #replacing empty cells with 0
```

```python
cong_codes = df['Cong Zip Codes'].tolist()    #making the congressional codes into a list
```

```python
voter_codes = df['Zip Code'].tolist()       #making the voter codes into a list
```

```python
def intersection(x, y):                      #function to check intersection betweem voter and cong codes
    lst3 = [value for value in cong_codes if value in voter_codes] 
    return lst3 
```

```python
intersecting_zipps = (intersection(cong_codes, voter_codes))  # saving the intersection of codes into a variable 
```

```python
intersecting_zipps
```

```python
final_correct_people = df[df['Zip Code'].isin(intersecting_zipps)]    # checking which intersecting zips are in the normal zip code list
```

```python
final_correct_people.drop(columns = ['Cong Zip Codes'], inplace = True) # dropping the cong zip code columns
```

```python
final_correct_people.head()                          # End up with a dataframe of purely the list of people that contain specific zip codes
```

```python
writer = pd.ExcelWriter('TSBwith2.xlsx')                #Code outputs dataframe to excel file in Jupyter 
# write dataframe to excel
final_correct_people.to_excel(writer) 
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
```
